```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import sys  

# Setting layout parameters
room_width, room_height = 10, 8  

def generate_random_obstacles(num_furniture=3):
    """Generate random furniture positions with fixed size (2x2) inside the room."""
    return [(random.randint(1, room_width-2), random.randint(1, room_height-2), 2, 2) for _ in range(num_furniture)]

def generate_random_person():
    """Generate a random person position inside the room (1x1 size)."""
    return [random.randint(1, room_width-2), random.randint(1, room_height-2)]

# Initializing robot position
robot_position = [0, 0]
collision_reason = None 

def check_collision():
    """Check if the robot collides with furniture or a person and exit the program if it does."""
    global collision_reason  # Allow modification of global variable
    
    # Checking collision with furniture
    for fx, fy, fw, fh in furniture_positions:
        if fx <= robot_position[0] <= fx + fw and fy <= robot_position[1] <= fy + fh:
            collision_reason = "furniture"
            print("\n ALERT: Robot collided with FURNITURE! ")
            visualize_room(robot_position, furniture_positions, painted_areas, person_position)
            exit_simulation()

    # Checking collision with person
    if robot_position == person_position:
        collision_reason = "person"
        print("\n ALERT: Robot collided with a PERSON! ")
        visualize_room(robot_position, furniture_positions, painted_areas, person_position)
        exit_simulation()

def exit_simulation():
    """Prints the final collision reason and exits the program."""
    if collision_reason:
        print(f"\n Simulation terminated due to collision with {collision_reason.upper()} ")
    sys.exit()

def paint_walls_random_order():
    """Move the robot randomly along the walls to paint them."""
    global furniture_positions, person_position, robot_position
    wall_positions = [(0, y) for y in np.arange(0, room_height, 0.5)] + \
                     [(x, room_height) for x in np.arange(0, room_width, 0.5)] + \
                     [(room_width, y) for y in np.arange(room_height, -0.5, -0.5)] + \
                     [(x, 0) for x in np.arange(room_width, -0.5, -0.5)]
    random.shuffle(wall_positions)  # Randomize movement order
    
    for x, y in wall_positions:
        furniture_positions = generate_random_obstacles()  # Randomly place furniture
        person_position = generate_random_person()  # Randomly place a person

        robot_position[0], robot_position[1] = x, y  # Move robot to new point
        check_collision()  # Check for collision and exit if needed

        if (x, y) not in painted_areas:
            painted_areas.append((x, y))
            visualize_room(robot_position, furniture_positions, painted_areas, person_position)
            time.sleep(0.1)  # Simulate painting time delay

def visualize_room(robot_position, furniture_positions, painted_areas, person_position):
    """Render the room with the robot, furniture, person, and painted walls."""
    plt.figure(figsize=(8, 6))
    plt.xlim(0, room_width)
    plt.ylim(0, room_height)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Robot Painting Walls Simulation")
    plt.xlabel("Room Width (m)")
    plt.ylabel("Room Height (m)")
    
    # Ploting painted walls
    for x, y in painted_areas:
        plt.plot(x, y, 'bo', markersize=2)  # Small blue dots for painted walls
    
    # Ploting furniture obstacles
    for fx, fy, fw, fh in furniture_positions:
        plt.gca().add_patch(plt.Rectangle((fx, fy), fw, fh, color='brown', label='Furniture'))
    
    # Ploting person (larger red marker)
    plt.plot(person_position[0], person_position[1], 'ro', markersize=14, label='Person')

    # Ploting robot position
    plt.plot(robot_position[0], robot_position[1], 'go', markersize=14, label='Robot')
    
    # Customizing legend to exclude duplicate labels
    handles, labels = plt.gca().get_legend_handles_labels()
    unique_labels = dict(zip(labels, handles))  # Remove duplicates
    plt.legend(unique_labels.values(), unique_labels.keys())
    
    plt.show()
    
    print(f"Furniture Positions: {furniture_positions}")
    print(f"Person Position: {person_position}")
    print(f"Painted Areas: {painted_areas}")

# Initializing painted areas and furniture
painted_areas = []
person_position = generate_random_person()  # Initial random person position

try:
    while True:
        paint_walls_random_order()
        time.sleep(5)  
except KeyboardInterrupt:
    print("\n Simulation interrupted by user.")
```
